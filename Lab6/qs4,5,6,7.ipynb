4.Calculate the correlation for the temperature vs ice-cream sales dataset.
Compare the covariance and correlation values.
Explain why correlation is easier to interpret.

```python
import pandas as pd
import numpy as np

temperature = [73, 68, 75, 79, 72, 69, 65, 77, 74, 78]
sales = [221, 180, 340, 360, 220, 150, 100, 400, 300, 410]

# Create the dataset
data = pd.DataFrame({
    'Temperature': temperature,
    'Sales': sales 
})

# mean
mean_temp = np.mean(data['Temperature'])
mean_sales = np.mean(data['Sales'])
print("Mean Temperature:", mean_temp)
print("Mean Sales:", mean_sales)

#coveriance
covariance = data['Temperature'].cov(data['Sales'])
print("Covariance:", covariance)

# correlation
correlation = data['Temperature'].corr(data['Sales'])
print("Correlation:", correlation)
```

5. Create two random numerical datasets with no clear relationship.
Compute their covariance and correlation.
Visualize them using a scatter plot and observe the pattern.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# generate the two random datasets
np.random.seed(42)
x= np.random.randn(50)
y= np.random.randn(50)

# claculate the coveriance

covariance = np.mean((x - np.mean(x)) * (y - np.mean(y)))
print("Covariance:", covariance)

# calculate the correlation
correlation = covariance / (np.std(x) * np.std(y))
print("Correlation:", correlation)

# scatter plot
plt.scatter(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Scatter Plot of x vs y")
plt.show()
```

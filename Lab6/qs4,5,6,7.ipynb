4.Calculate the correlation for the temperature vs ice-cream sales dataset.
Compare the covariance and correlation values.
Explain why correlation is easier to interpret.

```python
import pandas as pd
import numpy as np

temperature = [73, 68, 75, 79, 72, 69, 65, 77, 74, 78]
sales = [221, 180, 340, 360, 220, 150, 100, 400, 300, 410]

# Create the dataset
data = pd.DataFrame({
    'Temperature': temperature,
    'Sales': sales 
})

# mean
mean_temp = np.mean(data['Temperature'])
mean_sales = np.mean(data['Sales'])
print("Mean Temperature:", mean_temp)
print("Mean Sales:", mean_sales)

#coveriance
covariance = data['Temperature'].cov(data['Sales'])
print("Covariance:", covariance)

# correlation
correlation = data['Temperature'].corr(data['Sales'])
print("Correlation:", correlation)
```

5. Create two random numerical datasets with no clear relationship.
Compute their covariance and correlation.
Visualize them using a scatter plot and observe the pattern.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# generate the two random datasets
np.random.seed(42)
x= np.random.randn(50)
y= np.random.randn(50)

# claculate the coveriance

covariance = np.mean((x - np.mean(x)) * (y - np.mean(y)))
print("Covariance:", covariance)

# calculate the correlation
correlation = covariance / (np.std(x) * np.std(y))
print("Correlation:", correlation)

# scatter plot
plt.scatter(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Scatter Plot of x vs y")
plt.show()
```

6. Load a small dataset with height and weight values.
Calculate mean, variance, covariance, and correlation.
Plot the data using a seaborn jointplot.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# create a dataset
data = {
    'Height':[150,155,160,165,170,175,180],
    'Weight' :[50,55,60,65,70,75,80]
}

df = pd.DataFrame(data)

# calculate the mean
mean_height = df['Height'].mean()
mean_weight = df['Weight'].mean()
print("Mean Height:", mean_height)
print("Mean Weight:", mean_weight)

# clculat the veriance
var_height = df['Height'].var()
var_weight = df['Weight'].var()
print("Variance Height:", var_height)
print("Variance Weight:", var_weight)

# clculate the covariance
covariance = df['Height'].cov(df['Weight'])
print("Covariance:", covariance)

# calculate the correlation
correlation = df['Height'].corr(df['Weight'])
print("Correlation:", correlation)

# seaborn joinplot
sns.jointplot(x="Height", y="Weight", data=df, kind="scatter")
plt.show()
```

7. Create a dataset where one variable increases linearly and another increases randomly.
Compute covariance and correlation.
Plot both cases and compare the results.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Seed for reproducibility
np.random.seed(42)

# Case 1: Linear relationship

x_linear = np.arange(1, 21)
y_linear = 2 * x_linear   # linear increase

# Covariance and correlation (linear)
cov_linear = np.mean((x_linear - np.mean(x_linear)) *(y_linear - np.mean(y_linear)))
corr_linear = cov_linear / (np.std(x_linear) * np.std(y_linear))

print("LINEAR CASE")
print("Covariance:", cov_linear)
print("Correlation:", corr_linear)

# Plot linear case
plt.scatter(x_linear, y_linear)
plt.xlabel("X (Linear)")
plt.ylabel("Y (Linear)")
plt.title("Linear Relationship")
plt.show()
```

```python
x_random = np.arange(1, 21)
y_random = np.random.rand(20) * 40  # random values

# Covariance and correlation (random)
cov_random = np.mean((x_random - np.mean(x_random)) *(y_random - np.mean(y_random)))
corr_random = cov_random / (np.std(x_random) * np.std(y_random))

print("RANDOM CASE")
print("Covariance:", cov_random)
print("Correlation:", corr_random)

# Plot random case
plt.scatter(x_random, y_random)
plt.xlabel("X (Linear)")
plt.ylabel("Y (Random)")
plt.title("Random Relationship")
plt.show()
```

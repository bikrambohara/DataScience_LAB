Generate a dataset of daily temperatures and ice-cream sales. Find the covariance between temperature and sales. Plot the relationship using a scatter plot

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate dataset
np.random.seed(0)
days = 30
temperature = np.random.randint(20, 40, days)
sales = temperature * 5 + np.random.randint(-20, 20, days)

# Create DataFrame
data = pd.DataFrame({
    "Temperature": temperature,
    "IceCream_Sales": sales
})

# Calculate covariance
covariance = np.cov(temperature, sales)[0][1]
print("Covariance between Temperature and Ice Cream Sales:", covariance)

# Scatter plot
plt.figure()
plt.scatter(temperature, sales)
plt.xlabel("Temperature (Â°C)")
plt.ylabel("Ice Cream Sales")
plt.title("Temperature vs Ice Cream Sales")
plt.show()
```
